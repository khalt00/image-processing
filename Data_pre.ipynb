# code pre-processing

```python
width,height = 256,256
dim_resize = (width,height)


def cutImg(data,file,img_dir):
        img = cv2.imread(img_dir+ file)#đọc 1 ảnh
        image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        Ih,Iw,_ = image.shape # Lấy ra height, width của ảnh
        class_id, x_center, y_center, w, h = data # 0 0.500471 0.494063 0.999057 0.984375
        #sau khi label bằng labelImg, ta sẽ có annotation định dạng yolo
        #Những bước sau chuyển các định dạng yolo sang tọa độ thông thường để cắt ảnh
        x_center, y_center, w, h = float(x_center), float(y_center), float(w), float(h)
        x_center = round(x_center * Iw)
        y_center = round(y_center * Ih)
        w = round(w * Iw)
        h = round(h * Ih)
        x = round(x_center - w / 2)
        y = round(y_center - h / 2)
        imgCrop = image[y:y + h, x:x + w]
#         visualize1(imgCrop)
        #Xem ảnh được cắt
        new_image = cv2.resize(imgCrop,dim_resize)
#         visualize2(new_image)
        #Xem ảnh mới được resize
        image = Image.fromarray(new_image)
        return image


fold_dir = os.listdir('/home/khale/all/data/train/')

for animal in fold_dir:
    print(animal)
    img_dir = f"/home/khale/all/data/train/{animal}/"

    image_files= os.listdir(img_dir)

    i = 0
    for file in image_files:
        txt_file = file.split('.')[0]+'.txt'
        print(txt_file)
        txt = open(f'/home/khale/all/data_label/train/{animal}/' + txt_file)
        text_inside = txt.readlines()
        for lines in text_inside:
            i += 1
            temp = lines.split()
            print(temp)
            new_image = cutImg(temp,file,img_dir) #Gọi lại hàm bên trên
            image1 = new_image.save(f"/home/khale/data_256_test/train/{animal}/{i}.jpg" )

        
```
